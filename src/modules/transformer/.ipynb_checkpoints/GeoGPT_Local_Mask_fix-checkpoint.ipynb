```python
def get_xyt_by_pixel(pixel, frame_len = 256, camera_len = 30, h = 16, w = 16):
    # write a hand-craft-strategy!
    # first filter out camera
    if 256 <= pixel < 286 or 542 <= pixel < 572:
        return -1, -1, -1
    
    # first frame
    if 0 <= pixel < 256:
        t = 0
        x = pixel % w
        y = pixel // w
        return t, x, y
    
    # second frame
    if 286 <= pixel < 542:
        pixel = pixel - 286
        t = 1
        x = pixel % w
        y = pixel // w
        return t, x, y
    
    # second frame
    if 572 <= pixel < 828:
        pixel = pixel - 572
        t = 2
        x = pixel % w
        y = pixel // w
        return t, x, y
    
def judge_local(x, y, t, x_c, y_c, t_c, window = 1):
    # given a ref, judge if given point is in local
    # first all the (x,y) in window is in local
    if abs(x - x_c) <= window and abs(y - y_c) <= window:
        return True
    
    # then if (x,y) is not in local but in the same frame
    if t == t_c:
        if x == x_c or y == y_c:
            return True
        
    return False
```

```python
import torch
import matplotlib.pyplot as plt

def visual_mask(mask):
    plt.imshow(mask.numpy())
    plt.show()

block_size = 827
n_unmasked = 286

mask = torch.tril(torch.ones(block_size,
                             block_size))
visual_mask(mask)

mask[:n_unmasked, :n_unmasked] = 1
visual_mask(mask)
```

```python
visual_mask(mask)

for row in range(n_unmasked, block_size):
    t_c, x_c, y_c = get_xyt_by_pixel(row)
    for column in range(block_size):
        # first check if the pixel is unmask
        if mask[row, column] == 0:
            continue
        t, x, y = get_xyt_by_pixel(column)
        # then check if the pixel is camera
        if t == -1:
            continue
        # if (t,x,y) is not in local, set mask to 0
        if not judge_local(x, y, t, x_c, y_c, t_c):
            mask[row, column] = 0
        
visual_mask(mask)
```
